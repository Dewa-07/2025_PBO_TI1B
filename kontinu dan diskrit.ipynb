```python
import numpy as np
import matplotlib.pyplot as plt

# --- Sinyal Diskrit ---
n = np.arange(-10, 11, 1)   # index diskrit
x_n = np.sin(0.5 * n)       # fungsi sinyal diskrit

# --- Sinyal Kontinu ---
t = np.linspace(-10, 10, 500)  # waktu kontinu
x_t = np.sin(0.5 * t)          # fungsi sinyal kontinu

# --- Plot ---
plt.figure(figsize=(12,5))

# Plot sinyal diskrit
plt.subplot(1, 2, 1)
plt.stem(n, x_n, basefmt=" ")
plt.title("Sinyal Diskrit")
plt.xlabel("n")
plt.ylabel("x[n]")

# Plot sinyal kontinu
plt.subplot(1, 2, 2)
plt.plot(t, x_t)
plt.title("Sinyal Kontinu")
plt.xlabel("t")
plt.ylabel("x(t)")

plt.tight_layout()
plt.show()
```
